```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
adult = pd.read_csv("adult.csv")
```

```python
del adult['fnlwgt']
```

```python
adult.head()
```

```python
# transform numerical attributes into categorical ones
adult['age'] = pd.qcut(adult['age'], q=4)
adult['age'].value_counts(sort=False)
```

```python
adult['education-num'] = pd.qcut(adult['education-num'], q=4)
adult['education-num'].value_counts(sort=False)
```

```python
adult['capital-gain'] = pd.qcut(adult['capital-gain'], q=40, duplicates='drop')
adult['capital-gain'].value_counts(sort=False)
```

```python
adult['capital-loss'] = pd.qcut(adult['capital-loss'], q=70, duplicates='drop')
adult['capital-loss'].value_counts(sort=False)
```

```python
adult['hours-per-week'] = pd.qcut(adult['hours-per-week'], q=5, duplicates='drop')
adult['hours-per-week'].value_counts(sort=False)
```

```python
adult.head()
```

```python
adult.shape
```

```python
adult1 = adult.sample(frac=0.1, replace=False, random_state=1)
adult1.shape
```

```python
data_list = []
for row in range(0, 3256):
    data_list.append([str(adult1.values[row,column]) for column in range(0, 14)])

rules = apriori(data_list, min_support=0.25, min_confidence=0.25, min_length=2)
results = list(rules)
```

```python
print(results)
```

```python
#Put the results into a dataframe to have the rules listed in a table
df = pd.DataFrame(columns=('Items','Antecedent','Consequent','Support','Confidence','Lift'))

Support =[]
Confidence = []
Lift = []
Items = []
Antecedent = []
Consequent=[]

for RelationRecord in results:
    for ordered_stat in RelationRecord.ordered_statistics:
        Support.append(RelationRecord.support)
        Items.append(RelationRecord.items)
        Antecedent.append(ordered_stat.items_base)
        Consequent.append(ordered_stat.items_add)
        Confidence.append(ordered_stat.confidence)
        Lift.append(ordered_stat.lift)

df['Items'] = list(map(set, Items))
df['Antecedent'] = list(map(set, Antecedent))
df['Consequent'] = list(map(set, Consequent))
df['Support'] = Support
df['Confidence'] = Confidence
df['Lift']= Lift

#Sort resulted dataframe by Lift
df.sort_values(by ='Lift', ascending = False, inplace = True)
```

```python
 df.reset_index(drop=True, inplace=True)
```
