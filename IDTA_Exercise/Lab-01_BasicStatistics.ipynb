## Occupations vs Income

```python
occupation_income = pd.crosstab(adult["occupation"], adult["income"], normalize="index")

occupation_income.plot(kind="bar", figsize=(10,6))
plt.title("Occupation vs Income")
plt.xlabel("Occupation")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

## Education vs Income 

```python
education_income = pd.crosstab(adult["education"], adult["income"], normalize="index")

education_income.plot(kind="bar", stacked=True, figsize=(10,6))
plt.title("Education vs Income")
plt.ylabel("Proportion")
plt.xlabel("Education")
plt.xticks(rotation=45)
plt.legend(title="Income")
plt.show()
```

## Age vs Weekly working hour

```python
plt.scatter(adult['age'] , adult['hours-per-week'], s = 10 , marker="x" ,alpha=0.3)
plt.title( 'Age vs Working Hours per Week')
plt.xlabel( 'Age' ) 
plt.ylabel( 'Working Hours per Week')
plt.show() 
```
